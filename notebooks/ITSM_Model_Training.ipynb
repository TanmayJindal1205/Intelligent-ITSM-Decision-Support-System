```python
plt.figure(figsize=(12,6))

counts, bins, patches = plt.hist(
    filtered['Resolution_Minutes'],
    bins=15,
    edgecolor='black'
)

for count, patch in zip(counts, patches):

    if count > 0:

        plt.text(
            patch.get_x() + patch.get_width()/2,
            patch.get_height(),
            f'{int(count)}',
            ha='center',
            va='bottom',
            fontsize=9
        )

plt.title('Resolution Time Distribution (99% of Tickets)')
plt.xlabel('Resolution Time (Minutes)')
plt.ylabel('Number of Tickets')

plt.tight_layout()
plt.show()
```

**Key Findings**
* Most tickets are resolved within the first few hundred minutes.
* Resolution times are highly right-skewed, with a small number of tickets taking significantly longer than average.
* Only a small fraction of tickets contribute to the long resolution-time tail.

**Problem Identified**
* A small set of tickets experiences exceptionally long resolution times, indicating potential bottlenecks or delayed resolution processes.

**Recommendations**
* Investigate tickets with unusually high resolution times to identify common causes of delay.
* Establish escalation and monitoring mechanisms for tickets exceeding expected resolution thresholds.

**Insight**
* While overall ticket resolution is fast for most cases, a small number of delayed tickets contribute disproportionately to the total resolution workload.

# Technician Performance Analysis

```python
tech_summary = (
    df.groupby('Technician')
    .agg(
        Tickets=('RequestID','count'),
        Avg_Resolution=('Resolution_Minutes','mean')
    )
)
```

```python
# SEVERITY SCORES

priority_score = {
    'Severity - 1 (High Priority)': 4,
    'Severity - 2 (Medium Priority)': 3,
    'Severity - 3 (Low Priority)': 2,
    'Severity - 4 (Normal Priority)': 1
}

df['Severity_Score'] = df['Priority'].map(priority_score)


# TECHNICIAN PERFORMANCE TABLE

tech_summary = (
    df.groupby('Technician')
      .agg(
          Tickets_Handled=('RequestID', 'count'),
          Median_Resolution_Min=('Resolution_Minutes', 'median'),
          Avg_Severity=('Severity_Score', 'mean')
      )
)


# INDIVIDUAL RANKINGS

# Higher ticket load is better
tech_summary['Ticket_Load_Rank'] = (
    tech_summary['Tickets_Handled']
    .rank(method='dense', ascending=False)
)

# Lower median resolution time is better
tech_summary['Resolution_Rank'] = (
    tech_summary['Median_Resolution_Min']
    .rank(method='dense', ascending=True)
)

# Higher average severity handled is better
tech_summary['Severity_Rank'] = (
    tech_summary['Avg_Severity']
    .rank(method='dense', ascending=False)
)


# OVERALL SCORE

# Weights:
# 40% Ticket Load
# 40% Resolution Speed
# 20% Severity Complexity

tech_summary['Overall_Score'] = (
      0.4 * tech_summary['Ticket_Load_Rank']
    + 0.4 * tech_summary['Resolution_Rank']
    + 0.2 * tech_summary['Severity_Rank']
)


# FINAL RANKING

tech_summary = (
    tech_summary
    .sort_values('Overall_Score')
)

tech_summary['Final_Rank'] = range(
    1,
    len(tech_summary) + 1
)


# RESULTS

tech_summary = tech_summary[
    [
        'Final_Rank',
        'Tickets_Handled',
        'Median_Resolution_Min',
        'Avg_Severity',
        'Overall_Score'
    ]
]

tech_summary.head(20)
```

**Key Findings**
* Vinod Sharma, Naveen Kumar and Surender Singh are the top-performing technicians.
* Top-ranked technicians handle a high ticket volume while maintaining low resolution times.
* Most technicians handle tickets with similar severity levels (Avg Severity ≈ 2).
* DCO Hisar, Barcode and Rajender Prasad show noticeably higher resolution times than peers.

**Problem Identified**
* A few technicians have significantly longer resolution times despite handling substantial ticket volumes.
* Rankings for technicians with very low ticket counts (e.g., 1–13 tickets) may not be reliable.

**Recommendations**
* Analyze workflows of top performers and replicate best practices.
* Review cases handled by low-ranked technicians to identify bottlenecks.
* Exclude or separately evaluate technicians with very small ticket volumes.
* Provide targeted training or workload balancing for slower-performing teams.

**Insight**
* Technician performance differences are primarily driven by resolution efficiency rather than ticket complexity, as almost all technicians handle tickets with a similar average severity (~2).
* High-ranking technicians demonstrate the ability to manage large workloads while maintaining faster resolution times.

# **SLA** (Service Level Agreement) **Breach Analysis**

```python
sla_limits = {
    'Severity - 1 (High Priority)': 30,
    'Severity - 2 (Medium Priority)': 120,
    'Severity - 3 (Low Priority)': 240,
    'Severity - 4 (Normal Priority)': 240
}

df['SLA_Limit_Minutes'] = df['Priority'].map(sla_limits)
```

```python
df['SLA_Breached'] = (
    df['Resolution_Minutes'] > df['SLA_Limit_Minutes']
)
```

Response-time SLA was excluded from the analysis because the dataset does not contain ticket acknowledgement/first-response timestamps, making only resolution-time SLA measurable and verifiable.

```python
df.columns.tolist()
```

```python
df['SLA_Breached'].value_counts(dropna=False)
```

* **Total Tickets**    : 21,961
* **SLA Met**          : 18,243
* **SLA Breached**     : 3,718
* **SLA Compliance**   : 83.07%
* **SLA Breach Rate**  : 16.93%

```python
plt.figure(figsize=(6,4))

sns.countplot(x='SLA_Breached', data=df)

plt.title('Overall SLA Compliance')
plt.show()
```

**Key Findings**

* 83.1% of tickets were resolved within SLA.
* 16.9% of tickets breached SLA targets.

**Insight**

* SLA compliance is satisfactory, but reducing the breach rate remains a key opportunity for improving service performance.

```python
sla_rate = pd.crosstab(
    df['Priority'],
    df['SLA_Breached'],
    normalize='index'
) * 100

sla_rate[True].plot(
    kind='bar',
    figsize=(8,5)
)

plt.ylabel('SLA Breach %')
plt.title('SLA Breach Rate by Priority')
plt.show()
```

**Key Findings**

* High Priority tickets have the highest SLA breach rate (~50%).
* Medium and Normal Priority tickets show similar breach rates (~29%).
* Low Priority tickets have the lowest breach rate (~17%).

**Insight**

* SLA breach risk increases as ticket severity increases, with High Priority tickets being the most challenging to resolve within their stricter SLA targets.

```python
# Total tickets per category
total_tickets = df.groupby('Category').size()

# Breached tickets per category
breached_tickets = df[df['SLA_Breached']].groupby('Category').size()

# Breach %
breach_rate = (breached_tickets / total_tickets * 100).fillna(0)

# Ignore tiny categories (<30 tickets) to avoid misleading percentages
breach_rate_filtered = breach_rate[total_tickets >= 30]

# Top 15 worst categories by breach %
top_breach_rate = breach_rate_filtered.sort_values(ascending=False).head(15)

plt.figure(figsize=(10,6))
top_breach_rate.sort_values().plot(kind='barh')
plt.xlabel('SLA Breach Rate (%)')
plt.ylabel('Category')
plt.title('Top Categories by SLA Breach Rate')
plt.show()
```

```python
sla_analysis = pd.DataFrame({
    'Total_Tickets': total_tickets,
    'Breached_Tickets': breached_tickets,
    'Breach_Rate_%': breach_rate
}).fillna(0)

sla_analysis = sla_analysis[sla_analysis['Total_Tickets'] >= 30]

sla_analysis.sort_values(
    'Breach_Rate_%',
    ascending=False
).head(15)
```

**Key Findings**

* SOC Monitoring Offense (81.6%), Asset Refresh (76.1%), and Meeting Arrangement (75.5%) recorded the highest SLA breach rates.
* Asset Refresh generated the highest number of breached tickets (316), followed by User Administration (541) and PM Activity (1,134).
* High-volume categories such as PM Activity (28.1%) and User Administration (25.1%) also exhibit substantial breach rates.

Each candidate model is evaluated using stratified 5-fold cross-validation on the training dataset.
```python
import pandas as pd
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder

from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_validate

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTENC

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

```python
categorical_features = [

    'Priority',
    'Category',
    'Sub Category',
    'Department',
    'Group',
    'Site',
    'Request Type',
    'Created_Day',
    'Created_Month'

]

numerical_features = [

    'Created_Hour'

]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder

# Preprocessing pipeline

preprocessor = ColumnTransformer(

    transformers=[

        (
            'categorical',

            OrdinalEncoder(
                handle_unknown='use_encoded_value',
                unknown_value=-1
            ),

            categorical_features

        ),

        (
            'numerical',
            'passthrough',
            numerical_features
        )

    ]

)
```

```python
from imblearn.over_sampling import SMOTENC

categorical_indices = list(range(len(categorical_features)))

smote = SMOTENC(

    categorical_features=categorical_indices,
    random_state=42,
    k_neighbors=5

)
```

```python
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(

    n_splits=5,
    shuffle=True,
    random_state=42

)
```

```python
models = {

    'Logistic Regression':
        LogisticRegression(
            max_iter=1000,
            random_state=42
        ),

    'Decision Tree':
        DecisionTreeClassifier(
            random_state=42
        ),

    'Random Forest':
        RandomForestClassifier(
            n_estimators=200,
            random_state=42,
            n_jobs=-1
        ),

    'XGBoost':
        XGBClassifier(
            random_state=42,
            eval_metric='logloss'
        ),

    'LightGBM':
        LGBMClassifier(
            random_state=42,
            verbosity=-1
        ),

    'CatBoost':
        CatBoostClassifier(
            random_state=42,
            verbose=0
        )

}
```

```python
def evaluate_model(model):

    # Create pipeline
    pipeline = Pipeline(steps=[

        ('preprocessor', preprocessor),

        ('smote', smote),

        ('classifier', model)

    ])

    # Metrics to evaluate
    from sklearn.metrics import make_scorer

    scoring = {
        "accuracy": "accuracy",
        "precision": make_scorer(precision_score, pos_label=True),
        "recall": make_scorer(recall_score, pos_label=True),
        "f1": make_scorer(f1_score, pos_label=True),
        "roc_auc": "roc_auc"
    }

    # Perform Stratified 5-Fold Cross Validation
    scores = cross_validate(

        estimator=pipeline,

        X=X_train,

        y=y_train,

        cv=cv,

        scoring=scoring,

        n_jobs=-1

    )

    # Return average scores
    return {

        'Accuracy': scores['test_accuracy'].mean(),
        'Precision': scores['test_precision'].mean(),
        'Recall': scores['test_recall'].mean(),
        'F1-Score': scores['test_f1'].mean(),
        'ROC-AUC': scores['test_roc_auc'].mean()

    }
```

```python
results = []

for model_name, model in models.items():

    metrics = evaluate_model(model)

    metrics['Model'] = model_name

    results.append(metrics)
```

### 5.4 Baseline Model Comparison

Each candidate model is evaluated using **5-fold Stratified Cross-Validation** on the training dataset. The preprocessing pipeline, SMOTENC oversampling, and classifier are executed independently within each fold to prevent data leakage. Model performance is compared using Accuracy, Precision, Recall, F1-Score, and ROC-AUC, with **F1-Score** serving as the primary evaluation metric due to the imbalanced nature of the dataset.

```python
# Convert results into a comparison table

results_df = (
    pd.DataFrame(results)
      .set_index("Model")
      .sort_values(by="F1-Score", ascending=False)
      .round(4)
)

display(results_df)
```

Logistic Regression achieved the highest Recall (0.7431) but at the cost of substantially lower Precision (0.2427), resulting in the lowest F1-Score among all models.

**Key Findings**

## 6. Hyperparameter Tuning

The top-performing baseline models are further optimized using **GridSearchCV** with **5-fold Stratified Cross-Validation**. Hyperparameter optimization searches across predefined parameter combinations to identify the configuration that maximizes the **F1-Score**, ensuring a balanced trade-off between Precision and Recall for the imbalanced SLA breach prediction problem.

To improve code reusability and maintainability, a generic tuning function is implemented and reused for each selected model.

```python
from sklearn.model_selection import GridSearchCV
from imblearn.pipeline import Pipeline
```

```python
def tune_model(model, param_grid, model_name):
    """
    Performs GridSearchCV for a given model using
    preprocessing + SMOTENC + classifier pipeline.
    """

    pipeline = Pipeline([
        ("preprocessor", preprocessor),
        ("smote", smote),
        ("classifier", model)
    ])

    grid_search = GridSearchCV(
        estimator=pipeline,
        param_grid=param_grid,
        scoring="f1",
        cv=cv,
        n_jobs=-1,
        verbose=1
    )

    grid_search.fit(X_train, y_train)

    print("=" * 60)
    print(model_name)
    print("=" * 60)

    print("\nBest Parameters:")
    print(grid_search.best_params_)

    print("\nBest Cross Validation F1-Score:")
    print(round(grid_search.best_score_, 4))

    return grid_search
```

### 6.1 Random Forest Hyperparameter Tuning

Random Forest is optimized by tuning the number of trees, tree depth, and node splitting criteria. Grid Search evaluates each parameter combination using five-fold stratified cross-validation to identify the configuration that maximizes the F1-Score.

```python
rf_param_grid = {

    "classifier__n_estimators": [100, 200, 300],

    "classifier__max_depth": [None, 10, 20],

    "classifier__min_samples_split": [2, 5],

    "classifier__min_samples_leaf": [1, 2]

}
```

```python
rf_grid = tune_model(
    RandomForestClassifier(random_state=42),
    rf_param_grid,
    "Random Forest"
)

best_rf = rf_grid.best_estimator_
```

**Observation**

* Grid Search identified the optimal Random Forest configuration with **100 estimators**, **unrestricted tree depth**, **minimum split size of 5**, and **minimum leaf size of 1**. The tuned model achieved a **mean cross-validation F1-Score of 0.6453**, improving upon the baseline performance and demonstrating that moderate regularization of node splitting enhances generalization.

### 6.2 XGBoost Hyperparameter Tuning

XGBoost is optimized by tuning the number of boosting iterations, tree depth, learning rate, and subsampling ratio. These parameters control model complexity, learning behavior, and generalization performance. Grid Search with 5-fold Stratified Cross-Validation is used to identify the parameter combination that maximizes the F1-Score.

```python
xgb_param_grid = {

    "classifier__n_estimators": [100, 200],

    "classifier__max_depth": [4, 6, 8],

    "classifier__learning_rate": [0.05, 0.1],

    "classifier__subsample": [0.8, 1.0]

}
```

```python
xgb_grid = tune_model(

    XGBClassifier(
        random_state=42,
        eval_metric="logloss"
    ),

    xgb_param_grid,

    "XGBoost"

)

best_xgb = xgb_grid.best_estimator_
```

### 6.3 CatBoost Hyperparameter Tuning

CatBoost is optimized by tuning the number of boosting iterations, tree depth, and learning rate. These parameters directly influence the model's predictive capability and generalization performance. Grid Search with 5-fold Stratified Cross-Validation is employed to identify the optimal parameter configuration.

```python
cat_param_grid = {

    "classifier__iterations": [200, 300],

    "classifier__depth": [4, 6, 8],

    "classifier__learning_rate": [0.05, 0.1]

}
```

```python
cat_grid = tune_model(

    CatBoostClassifier(
        random_state=42,
        verbose=0
    ),

    cat_param_grid,

    "CatBoost"

)

best_cat = cat_grid.best_estimator_
```

## 9) SHAP Explainability

While Random Forest feature importance provides a global ranking of predictor variables, it does not indicate how individual features influence model predictions. To improve model interpretability, **SHapley Additive exPlanations (SHAP)** are employed.

SHAP explains the contribution of each feature to the model's predictions using concepts from cooperative game theory. Unlike traditional feature importance, SHAP provides both **global explanations** (overall feature influence across the dataset) and **local explanations** (feature contributions for an individual prediction), making it a widely adopted explainability technique in modern machine learning applications.

```python
!pip install shap -q
```

```python
# Transform the test data using the fitted preprocessor
X_test_transformed = final_model.named_steps["preprocessor"].transform(X_test)

print(type(X_test_transformed))
print(X_test_transformed.shape)
```

```python
import shap

# Extract trained Random Forest
rf_model = final_model.named_steps["classifier"]

# Create SHAP explainer
explainer = shap.TreeExplainer(rf_model)

# Compute SHAP values
shap_values = explainer.shap_values(X_test_transformed)
```

```python
print(type(shap_values))

if isinstance(shap_values, list):
    print("Length:", len(shap_values))
else:
    print("Shape:", shap_values.shape)
```

### 9.1 SHAP Summary Plot

The SHAP Summary Plot provides a global explanation of the trained Random Forest model by illustrating both the importance of each feature and the direction of its influence on the model's predictions. Each point represents an individual ticket, while the color indicates the feature value. Features are ordered according to their overall impact on the model.

```python
# SHAP Summary Plot (SLA Breached class)

shap.summary_plot(
    shap_values[:, :, 1],
    X_test_transformed,
    feature_names=X_test.columns,
    show=True
)
```

**Observation**

- The SHAP Summary Plot provides a global explanation of the Random Forest model by illustrating both the importance of each feature and the direction of its influence on SLA breach predictions. Based on the mean absolute SHAP values, **Category, Request Type, Created_Month, Sub Category, and Department** are the most influential predictors, whereas **Priority** and **Site** contribute comparatively less to the model's decisions.

- The horizontal spread of SHAP values indicates the magnitude of each feature's influence on the prediction. Features with a wider distribution have a stronger impact on the model's output, while those concentrated near zero contribute relatively little. The color gradient represents the encoded feature values, allowing visualization of how different feature values are associated with positive or negative contributions to the model's predictions.

- Overall, the SHAP analysis indicates that **ticket category, request characteristics, and temporal attributes** are the primary contributors to the model's predictions, providing a transparent and interpretable explanation of the model's decision-making process.

### 9.3 SHAP Waterfall Plot

While the SHAP Summary Plot and SHAP Feature Importance Bar Plot provide global explanations of the Random Forest model, the SHAP Waterfall Plot provides a **local explanation** for an individual prediction. It illustrates how each feature contributes to the prediction by starting from the model's baseline output and sequentially adding the positive and negative contributions of individual features until the final prediction is obtained.

To demonstrate the model's decision-making process for both classes, waterfall plots are presented for one correctly predicted **SLA Met** ticket and one correctly predicted **SLA Breached** ticket.

```python
# Predict on test data
y_pred = final_model.predict(X_test)

# Correctly predicted SLA Met
sla_met_idx = np.where((y_test == 0) & (y_pred == 0))[0][0]

# Correctly predicted SLA Breached
sla_breached_idx = np.where((y_test == 1) & (y_pred == 1))[0][0]

print("SLA Met Sample :", sla_met_idx)
print("SLA Breached Sample :", sla_breached_idx)
```

```python
def create_feature_labels(sample_index):

    row = X_test.iloc[sample_index]

    labels = []

    for col in X_test.columns:
        labels.append(f"{col}: {row[col]}")

    return labels
```

#### 9.3.1 Correctly Predicted SLA Met Ticket

The following SHAP Waterfall Plot explains the prediction for a correctly classified ticket that is predicted to **meet the SLA**.

```python
feature_labels = create_feature_labels(sla_met_idx)

explanation = shap.Explanation(

    values=shap_values[sla_met_idx,:,1],

    base_values=explainer.expected_value[1],

    data=X_test_transformed[sla_met_idx],

    feature_names=feature_labels

)

shap.plots.waterfall(explanation)
```

- For the selected ticket, the baseline model output (**E[f(X)] = 0.50**) is progressively reduced to a final prediction of **f(x) = 0.02**, indicating a very low probability of an SLA breach. Consequently, the Random Forest model correctly classifies the ticket as **SLA Met**.

- The largest negative contributions are provided by **Request Type (Incident)**, **Category (Printers / Scanners)**, **Department (Finance & Accounts)**, **Created Month (July)**, and **Sub Category (Printer Hardware)**. These features collectively reduce the predicted probability of an SLA breach, while the remaining features contribute only marginally to the final prediction.

- Overall, the SHAP Waterfall Plot demonstrates how multiple feature contributions combine to produce a transparent and interpretable prediction for an individual ticket, highlighting the reasoning behind the Random Forest model's decision.

```python
feature_labels = create_feature_labels(sla_breached_idx)

explanation = shap.Explanation(

    values=shap_values[sla_breached_idx,:,1],

    base_values=explainer.expected_value[1],

    data=X_test_transformed[sla_breached_idx],

    feature_names=feature_labels

)

shap.plots.waterfall(explanation)
```
